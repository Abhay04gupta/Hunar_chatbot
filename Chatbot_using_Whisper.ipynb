```python
# @title Taking Audio
path="/content/ENGFND1_0368.mp3"
import IPython.display as display
display.Audio(path, autoplay=True)  #/content/ENGFND1_0368.mp3 : path of audio
```

```python
# @title Creating pipeline to convert audio to text
pipe = pipeline("automatic-speech-recognition", model="openai/whisper-small.en",device=0) #device=0


question=pipe(path)
question=question["text"]
print(question)
```

```python
# @title Model to generate response to query
model_cpt = "tiiuae/falcon-7b-instruct"
huggingface_api_token = "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"

llm = HuggingFaceHub(
    repo_id=model_cpt,
    model_kwargs={"temperature": 0.7, "max_new_tokens": 500},
    huggingfacehub_api_token=huggingface_api_token
)

template="""Question: {question}
Answer:"""

prompt=PromptTemplate(template=template,input_variables=['question'])

chain=LLMChain(prompt=prompt,llm=llm)

out=chain.run(question)
answer=out.split("Answer:")[1].replace("User","")
print(f"Question: {question}")
print(f"Answer: {answer}")

# Create a gTTS object
tts = gTTS(text=answer, lang='en')

# Save the audio to a file
tts.save("output.mp3")

import IPython.display as display
display.Audio("/content/output.mp3", autoplay=True)
```
